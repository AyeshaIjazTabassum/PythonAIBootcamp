# **Basic Operators**

1. **Arithmetic operators:** +, -, *, / etc.

2. **Assignment operators:** =, +=, -= etc.

3. **Comparison operators:** ==, >, >=, <, != etc.

4. **Logical operators:** and, or, not.









```python
#Arithmetic operators
a = 10
b = 3

print(a + b)  # 13
print(a - b)  # 7
print(a * b)  # 30
print(a / b)  # 3.333...
```

```python
#Assignment operators
x = 10
print("x =", x)

# Add and assign
x += 5   # x = x + 5
print("x after += 5:", x)

# Subtract and assign
x -= 3   # x = x - 3
print("x after -= 3:", x)

# Multiply and assign
x *= 2   # x = x * 2
print("x after *= 2:", x)
```

```python
#Comparison operators
a = 10
b = 20

print("a == b:", a == b)   # equal
print("a != b:", a != b)   # not equal
print("a > b:", a > b)     # greater than
print("a < b:", a < b)     # less than
print("a >= b:", a >= b)   # greater or equal
print("a <= b:", a <= b)   # less or equal
```

```python
#Logical operators
x = True
y = False

print("x and y:", x and y)   # both True
print("x or y:", x or y)     # at least one True
print("not x:", not x)       # reverses the value
```

**String Functions**

**len()** : counts characters

**capitalize()**:	 Converts the first character to upper case

**count()**:	Returns the number of times a specified value occurs in a string

**endswith()**:	Returns true if the string ends with the specified value

**startswith()**:	Returns true if the string starts with the specified value

**isalnum()**: Returns True if all characters in the string are alphanumeric

**lower()**:	Converts a string into lower case

**upper()**: Converts a string into upper case

```python
text = "Hello Python"
print(len(text))      # Counts characters including spaces
```
