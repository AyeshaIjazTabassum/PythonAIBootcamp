# Linear Regression

```python
from sklearn import datasets
```

```python
# 1. load dataset
diabetes = datasets.load_diabetes()
```

```python
diabetes
```

```python
# 2. create x and y metrices
x = diabetes.data
y = diabetes.target
```

```python
print(f"x: {x} \n \n y: {y}")
```

```python
x.shape, y.shape
```

```python
# 3. split into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)  # 80 - 20 split
```

```python
# 4. select model
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
# 5. train model
model.fit(x_train, y_train)
```

```python
# 6. predict
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
# 7. evaluate
from sklearn.metrics import mean_squared_error, r2_score, root_mean_squared_error
mse = mean_squared_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
mse, rmse, r2
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create an index for the samples
indices = np.arange(len(y_test))

plt.figure(figsize=(10,6))
plt.plot(indices, y_test, label='Actual (y_test)', color='blue', marker='o')
plt.plot(indices, y_pred, label='Predicted (y_pred)', color='orange', marker='x')

plt.xlabel("Sample Index")
plt.ylabel("Target Value")
plt.title("Actual vs Predicted Values")
plt.legend()
plt.grid(True)
plt.show()
```
