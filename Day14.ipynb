#Artificial Neural Network (ANN)

```python
# dataset
from sklearn.datasets import fetch_openml

dataset = fetch_openml("olivetti_faces", version=1)
```

```python
#2.
x = dataset.data
y = dataset.target
```

```python
x.shape, y.shape
```

```python
#3. split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
#4. model
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(hidden_layer_sizes=(128, 64, 32),
                      activation = 'relu',
                      solver = 'adam',
                      max_iter= 500,
                      random_state = 42)
```

```python
#5.
model.fit(x_train, y_train)
```

```python
#6.
y_pred = model.predict(x_test)
```

```python
# 7. evaluate
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
acc
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

# **Self Practice Question**

### ANN for Regression

```python
from sklearn.datasets import fetch_openml
dataset = fetch_openml("california_housing", version = 7)
```

```python
x = dataset.data
y = dataset.target
```

```python
x.shape , y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.neural_network import MLPRegressor
model = MLPRegressor(hidden_layer_sizes=(128, 64, 32),
                      activation = 'relu',
                      solver = 'adam',
                      max_iter= 500,
                      random_state = 42)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
mse
```
