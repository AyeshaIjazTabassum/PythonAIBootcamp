#**Computer Vision**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load image (BGR format)
img = cv2.imread("sample.png")

# Convert BGR to RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)    # cvtColor = convert color

# Display image
plt.imshow(img)
plt.axis("off")   # matplotlib command to hide axis line
```

```python
# Inspect image
print("Shape:", img_rgb.shape)
print("Data type:", img_rgb.dtype)
```

```python
# Resize image
resized_img = cv2.resize(img_rgb, (224, 224))  #parameters = width, height

plt.imshow(resized_img)
plt.axis("off")

print("Resized Shape:", resized_img.shape)
```

```python
# Convert to Grayscale
gray_img = cv2.cvtColor(resized_img, cv2.COLOR_RGB2GRAY)

plt.imshow(gray_img, cmap="gray")
plt.axis("off")

print("Gray Shape:", gray_img.shape)
```

```python
# Normalize pixel values
normalized_img = resized_img / 255.0

print("Min pixel value:", normalized_img.min())
print("Max pixel value:", normalized_img.max())
```

```python
plt.imshow(normalized_img)
```

```python
# Crop Image (center crop)
h, w, _ = resized_img.shape

crop = resized_img[h//4:3*h//4, w//4:3*w//4]

plt.imshow(crop)
plt.axis("off")

print("Cropped Shape:", crop.shape)
```

```python
# Padding Image
padded_img = cv2.copyMakeBorder(
    resized_img,
    20, 20, 20, 20,
    cv2.BORDER_CONSTANT,
    value=[0, 0, 0]
)

plt.imshow(padded_img)
plt.axis("off")

print("Padded Shape:", padded_img.shape)
```

```python
# Rotate Image
(h, w) = resized_img.shape[:2]
center = (w // 2, h // 2)

matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_img = cv2.warpAffine(resized_img, matrix, (w, h))

plt.imshow(rotated_img)
plt.axis("off")
```

```python
# Flip Image
flipped_img = cv2.flip(resized_img, 1)  # horizontal flip

plt.imshow(flipped_img)
plt.axis("off")
```

```python
# Adjust Brightness
bright_img = cv2.convertScaleAbs(resized_img, alpha=1, beta=50)

plt.imshow(bright_img)
plt.axis("off")
```

```python
# Apply Gaussian Blur
blurred_img = cv2.GaussianBlur(resized_img, (7, 7), 0)

plt.imshow(blurred_img)
plt.axis("off")
```
