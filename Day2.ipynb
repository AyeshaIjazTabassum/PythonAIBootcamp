# **What Are Data Structures?**

Data structures are special containers that **store multiple** values in Python.
They help us:



*    Organize data


*   Modify and manage large amounts of information easily



*    Access data faster



Today, we learn 3 important data structures:


*   **Lists**

*   **Tuples**
*   **Dictionaries**








# **Lists**

A list is an **ordered collection** of items.
It is **changeable**, which means we can add, remove, or update values.

**✔ Characteristics**


*   Written in square brackets → [ ]
*   Items are indexed (starting from 0)


*   Supports adding/removing items



```python
fruits = ["apple", "banana", "mango"]
marks = [85, 90, 78]
mix = ["asma", 20, True]
```

# **List Operations**
 **1. Accessing elements**


```python
fruits = ["apple", "banana", "mango"]
print(fruits[0])   # apple
print(fruits[2])   # mango
```

**2. Changing element**

```python
fruits[1] = "orange"
print(fruits)  # ["apple", "orange", "mango"]
```

**3. Adding items**

```python
fruits.append("grapes")
print(fruits)
```

**4. Removing items**

```python
fruits.remove("apple")
print(fruits)
```

**5. List length**

```python
print(len(fruits))
```

# **3. Tuples**

A tuple is similar to a list but **cannot be changed**.
It is immutable.

**✔ Characteristics**



*   Written in parentheses → ( )

*   Ordered
*   Cannot modify values





**Accessing tuple elements**

```python
colors = ("red", "green", "blue")
print(colors[0])  # red

thistuple = ("apple", "banana", "cherry")
print(thistuple[-1])
```

**Add Item**

```python
thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.append("orange")
thistuple = tuple(y)
print(thistuple)
```

**Update tuple**

```python
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)

print(x)
```

**Tuples cannot be changed**

```python
colors[1] = "yellow"
# ❌ This will give an error
```

# **4. Dictionaries**

Dictionaries store data in key : value pairs.
Each item has a name (key) and a value.

**✔ Characteristics**



*   Data is stored as key-value pairs






*   Written in curly braces → { }




*   Keys must be unique
*   Values can be changed



```python
student = {
    "name": "Asma",
    "age": 20,
    "grade": "A"
}
```

**Dictionary Operations**

**Accessing value**

```python
student = {"name": "Asma", "age": 20}
print(student["name"])  # Asma
```

**Changing values**

```python
student["age"] = 21
print(student)
```

**Adding new key-value**

```python
student["city"] = "Bahawalpur"
print(student)
```

**Removing key-value**

```python
student.pop("age")
print(student)
```

**Lists** → Changeable, ordered → ["apple", "banana"]

 **Tuples** → Unchangeable, ordered → ("red", "green")

**Dictionaries** → Key-value pairs → {"name": "Asma", "age": 20}

# **Sets**

A set in Python is a **collection of unique items**.

It is mostly used when you want to remove duplicates  or check membership quickly.

**✔ Characteristics of Sets:**

*   Written in curly braces → { }

*   No duplicates allowed
*   Unordered (items have no index)


*   Changeable (you can add/remove items)

```python
numbers = {1, 2, 3, 4}
```

```python
duplicates = {1, 2, 2, 3}
print(duplicates)
# Output: {1, 2, 3}
# Python automatically removes duplicate "2"
```

**Set Operations**

**1. Creating a Set**

```python
my_set = {10, 20, 30}
print(my_set)
```

**2. Adding Items**

```python
fruits = {"apple", "banana"}
fruits.add("mango")
print(fruits)
```

 **3. Removing Items**

```python
fruits.remove("banana")
print(fruits)
```

**4. Checking If Value Exists**

```python
numbers = {5, 10, 15, 20}
print(10 in numbers)   # True
print(50 in numbers)   # False
```

**5. Set Length**

```python
print(len(numbers))
```

**6. Removing Duplicates Using Set**

```python
data = [1, 2, 2, 3, 3, 3, 4]
unique_data = set(data)
print(unique_data)  # {1, 2, 3, 4}
```

**7. Union (Combine Sets)**

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
# Output: {1, 2, 3, 4, 5}
```

**8. Intersection (Common Values)**

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.intersection(b))
# Output: {2, 3}
```
