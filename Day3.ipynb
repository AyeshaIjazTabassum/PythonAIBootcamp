**Using if-elif-else (Multiple Conditions)**


```python
marks = int(input("Enter your marks: "))

if marks >= 90:
    print("Grade: A üåü")
elif marks >= 80:
    print("Grade: B üëç")
elif marks >= 70:
    print("Grade: C üôÇ")
else:
    print("Grade: Fail üò¢")
```

**Nested if Example (Important)**



‚úÖ Real-life example: If it's a holiday, check whether it's raining or sunny.



```python
is_holiday = input("Is it a holiday? (yes/no): ").lower()

if is_holiday == "yes":
    weather = input("What's the weather like? (rainy/sunny): ").lower()

    if weather == "rainy":
        print("Stay home and watch a movie üé¨")
    elif weather == "sunny":
        print("Go out and enjoy the sun ‚òÄÔ∏è")
    else:
        print("Enjoy your day however you like! üòé")
else:
    print("Get ready for work or school! üè´üíº")
```

# **For Loop**

A loop l**ets you repeat a block of code** as many times as needed ‚Äî just like when you say:

üß† ‚ÄúKeep eating until the plate is empty.‚Äù

```python
fruits = ["üçé Apple", "üçå Banana", "üçá Grape", "üçâ Watermelon"]

print("Fruit salad time!")
for fruit in fruits:
    print(f"- {fruit}")
```

# **While Loop**

**Basic while**

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

**Using break**

```python
i = 1

while True:
    print(i)
    if i == 3:
        break
    i += 1
```

**Using continue**

```python
i = 0

while i < 5:
    i += 1
    if i == 3:
        continue
    print(i)
```

```python
import random

secret = random.randint(1, 20)
guess = None
attempts = 0

print("Guess the secret number between 1 and 20!")

while guess != secret:
    guess = int(input("Your guess: "))
    attempts += 1
    if guess < secret:
        print("Too low! Try again ‚¨ÜÔ∏è")
    elif guess > secret:
        print("Too high! Try again ‚¨áÔ∏è")

print(f"You got it in {attempts} attempts! üéâ")
```

**Nested Loops ‚Äì Loops Inside Loops**



Great for tables, grids, or patterns. Example: Multiplication table, pattern printing:

```python
for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i*j}", end=" ")
    print()  # new line after inner loop
```

**üí• Fun Coding Challenge for Students:**



Write a program that checks if a number is positive, negative, or zero, and inside it, check if the number is even or odd (nested condition).

```python
num = int(input("Enter a number: "))

if num > 0:
    print("The number is Positive.")
    if num % 2 == 0:
        print("It's also Even.")
    else:
        print("It's also Odd.")
elif num < 0:
    print("The number is Negative.")
else:
    print("The number is Zero.")
```
