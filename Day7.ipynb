ðŸ“Œ **6. Convert Columns to Correct Types (if needed)**

```python

categorical_cols = [
    "gender","payment_method","city","product_category",
    "device_type","subscription_type","has_promo_code","churn"
]

for col in categorical_cols:
    df[col] = df[col].astype("category")

df.dtypes
```


Many columns represent categories, not numbers

Converting helps with EDA + ML later

# ðŸ“Š 7. Univariate Analysis

```python
df.hist(figsize=(18, 14), bins=20)
plt.show()
```

# â–¶ Countplots of Categorical Columns

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x="gender")
plt.show()
```

```python
cat_cols = ["gender","payment_method","city","product_category",
            "device_type","subscription_type","has_promo_code","churn"]

for col in cat_cols:
    plt.figure(figsize=(12,6))
    sns.countplot(data=df, x=col)
    plt.title(f"Countplot â€“ {col}")
    plt.xticks(rotation=45)
    plt.show()
```

# ðŸ“Š 8. Bivariate Analysis

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="income", y="purchase_amount", hue="gender")
plt.show()
```

6. Countplot vs Barplot

```python
fig, axes = plt.subplots(1, 2, figsize=(12,4))

sns.countplot(x="gender", data=df, ax=axes[0]) #countplot â†’ counts frequency
axes[0].set_title("Gender Count")

sns.barplot(x="gender", y="income", data=df, ax=axes[1]) #barplot â†’ shows mean of numeric column
axes[1].set_title("Average Income by Gender")

plt.show()
```
